```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Figura 3.8 - Transformações de potência (gama)

```python
def gamma_transform(img, c, gamma):
    return c*(img**gamma)

img = cv2.imread('chapter_3/Fig3.08(a).jpg',0)

img_b = gamma_transform(img, 1, 0.6).astype(np.uint8)
img_c = gamma_transform(img, 1, 0.4).astype(np.uint8)
img_d = gamma_transform(img, 1, 0.3).astype(np.uint8)

fig, axs = plt.subplots(2, 2)
axs[0, 0].imshow(img, cmap='gray', vmin=0, vmax=255)
axs[0, 1].imshow(img_b, cmap='gray')
axs[1, 0].imshow(img_c, cmap='gray')
axs[1, 1].imshow(img_d, cmap='gray')

for ax in axs.flat:
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.label_outer()


fig.subplots_adjust(left=0, right=1, top=2, bottom=0)
fig.savefig("chapter_3/answers/Fig38.png",bbox_inches='tight')
```

# Figura 3.12 - Fatiamento de níveis de intensidade.

```python
def img_level_b(min_lvl, max_lvl, realce, img):
    img_out = img;
    
    rows, columns = np.shape(img) 

    for i in range(rows):
        for j in range(columns):
            pixel = img_out[i,j];
            
            if pixel >= min_lvl and  pixel <= max_lvl:
                img_out[i,j] = realce;
            else:
                img_out[i,j] = 0;
            
    return img_out
    
def img_level_c(min_lvl, max_lvl, realce, img):
    img_out = img;
    
    rows, columns = np.shape(img) 

    for i in range(rows):
        for j in range(columns):
            pixel = img_out[i,j];
            
            if pixel >= min_lvl and  pixel <= max_lvl:
                img_out[i,j] = realce;
            elif pixel < min_lvl:
                img_out[i,j] = 0;
    
    return img_out

A = cv2.imread('chapter_3/Fig3.12(a)(kidney).tif',0).astype(np.uint8)

# Efetuando o fatiamento de niveis de intensidade modo 1
B = img_level_b(150, 255, 235, A).astype(np.uint8)

# Efetuando o fatiamento de niveis de intensidade modo 2
C = img_level_c(110, 145, 20, A).astype(np.uint8)


fig, axs = plt.subplots(1, 3)
axs[0].imshow(A, cmap='gray')
axs[1].imshow(B, cmap='gray')
axs[2].imshow(C, cmap='gray')

for ax in axs.flat:
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.label_outer()


fig.subplots_adjust(left=0, right=1, top=2, bottom=0)
fig.savefig("chapter_3/answers/Fig312.png",bbox_inches='tight')
```
